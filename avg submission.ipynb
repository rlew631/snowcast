The results of this notebook produced a 700 row long file which was not accepted for submission. The contest is expecting a 9066 row long file which uses the more verbose alphanumeric cell_id format found in the submission_format.csv file

```python
import pandas as pd
import datetime
```

```python
historic_data = pd.read_csv('all_days_training.csv')
sample_submission = pd.read_csv('submission_format.csv')
```

```python
#first day is 2013-01-01
# last day is 2019-12-31
first = datetime.date.fromisoformat('2013-01-01')
last = datetime.date.fromisoformat('2019-12-31')
```

```python
avg_submission = historic_data[['Unnamed: 0']].rename(columns={'Unnamed: 0': 'cell_id'})
```

```python
year_range = list(range(2013,2020))
```

```python
for val in sample_submission.columns[1:]:
#     print(val[5:])
#     2013 to 2019 avgs
    avg = historic_data[[ '2013-'+val[5:] ]].squeeze()
    for year in year_range[1:]:
        colname = str(year) + '-' + val[5:]
        avg = avg + historic_data[[colname]].squeeze()
    avg = avg / 7
    avg_submission[[val]] = avg
```

```python
avg_submission.set_index('cell_id', inplace=True)
avg_submission.head()
```

```python
avg_submission.to_csv('first_submission.csv')
```
