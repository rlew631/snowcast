# Resampling (interpolating)

```python
train_features_df.columns[1:32].resample('D', axis=1).mean()
```

```python
train_features_df.iloc[:,1:32]#.resample('D', axis=1).mean()
```

```python
date_labels = train_features_df.columns.values[1:]
for i, date in enumerate(date_labels):
    date_labels[i] = pd.to_datetime(date, format='%d%b%Y')
train_features_df.columns.values[1:] = date_labels
```

```python
train_features_df.iloc[:,1:32]
```

```python
train_features_df = train_features_df.iloc[:,1:32]
```

```python
date_labels = train_features_df.columns.values
for i, date in enumerate(date_labels):
    print(type(date))
```

```python
train_features_df.head(5)
```

```python
train_features_df = train_features_df.T
train_features_df.index = pd.to_datetime(train_features_df.index)
```

```python
train_features_df = train_features_df.resample('D').mean()
train_features_df
```

```python
train_features_df = train_features_df.interpolate()
train_features_df
```
